### Load the Data

```python
def load_data(filename) :
    with open(filename,"r") as f :
        return json.load(f)
```

## Cleaning And Structuring Data

### Problems In Data

1. Inconsistency In Data types of Rating
2. Missing Data
3. Duplicate Data 

```python
# Remove In consistency Of Rating
```

```python
def clean_data(data) : 
    map = {"one" : 1,"two" : 2,"three" : 3,"four" : 4,"five" : 5}
    cleaned_data = []
    name = []
    for dict in data:
        # clean Rating
        rat = dict.get("rating").strip().lower()
        if(rat in map):
            rat = map.get(rat)
        dict["rating"] = float(rat);

        # Handling Missing Data
        dict["age"] = dict.get("age")

        # remove duplicate 
        if(dict["name"] not in name):
            cleaned_data.append(dict)
        name.append(dict["name"])
    return cleaned_data
```

```python
data = load_data("data.json")
```

```python
data = clean_data(data)
```

### Getting Meaningful Insights from Data 

```python
# What is the Average Rating Of the Product
def get_avg_rating(data):
    avg_rating = 0
    for user in data:
        avg_rating+=(user.get("rating"))
    return avg_rating / (len(data))
```

```python
get_avg_rating(data)
```

```python
def get_percent(data):
    less_than_three_rating = 0

    for user in data:
        if(user.get("rating") < 3):
            less_than_three_rating+=1

    return (less_than_three_rating / len(data)) * 100
```

```python
get_percent(data)
```

### Implement Basic Recommendation System

```python
def get_recommendations(data):
    recommend_dict = {}
    for user in data :
        if(user.get("rating") >= 4):
            recommend_dict.update({user["name"] : "Apple"})
        else:
            recommend_dict.update({user["name"] : "samsung"})
    return recommend_dict
```

```python
get_recommendations(data)
```
