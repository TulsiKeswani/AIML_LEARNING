```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr ** 2)
print(arr + 10)
```

```python
# Broadcasting

arr1 = np.array([[1,2],[3,4],[5,6]])
arr2 = np.array([4])

print(arr1.shape)
print(arr2.shape)
print(arr1 * arr2)
```

```python
# Not Allowed broadcasting
# # arr1 = np.array([[1,2,3],[4,5,6]])  
# # arr2 = np.array([[1,2],[3,4]])
# print(arr1.shape)
# print(arr1.shape)
# print(arr1 * arr2)

arr1 = np.array([[1,2,3],[4,5,6]])  
arr2 = np.array([[2],[4]])
print(arr1.shape)
print(arr1.shape)
print(arr1 * arr2)
```

```python
# Standard Normalization
arr = np.array([[1,2,3],[4,5,6]])
std_dev = np.sqrt(np.sum(((arr - np.mean(arr)) ** 2 )) / arr.size)

# Normalization
norm_arr = (arr - np.mean(arr)) / std_dev
```

```python
print(std_dev)
print(np.std(arr))
print(norm_arr)
```

```python
print(np.mean(norm_arr))
```
