```python
import numpy as np
```

## Creating Numpy Array Using Python list

```python
# Creating 1D Array

arr1 = np.array([1,2,3,4,5])
print(arr1,type(arr1),arr1.dtype,arr1.shape)

arr2 = np.array([1,2,3,4,5,"Tulsi"])
print(arr2,type(arr2),arr2.dtype,arr2.shape)

arr3 = np.array([1,2,3,4,5,{"name" : "Tulsi"}])
print(arr3,type(arr3),arr3.dtype,arr3.shape)

# Creating 2D Array

arr_2D = np.array([[1,2,3,4],[5,6,7,10]])
print(arr_2D,arr_2D.shape)
```

## Creating Numpy array from scratch

```python
arr1 = np.zeros((3,4),dtype="int64")  # Fill with zeores
print(arr1,"\n",arr1.shape)

arr2 = np.ones((4,4),dtype="int64")  # Fill with Ones
print(arr2,"\n",arr2.shape)

arr3 = np.full((4,3),100,dtype="int64") #Fill with Specified Value
print(arr3,"\n",arr3.shape)

arr4 = np.eye(4,dtype=int)  # Creates Identity Matrix
print(arr4,arr4.shape)

arr5 = np.arange(1,10,2)   # Similar Like Python Range
print(arr5,arr5.shape)

arr6 = np.linspace(1,100,10,dtype=int) # create array with equally spaced values
print(arr6)
```
