```python
import numpy as np
```

```python
# Aggregation Functions

arr = np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr)) #return index of minimum value
print(np.argmax(arr))  #return index of maximum value
print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))

# varian == square(std)
print(np.square(np.std(arr)))
print(np.var(arr))
```

```python
# Power Functions

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
# Log
print(np.log(arr))
print(np.log2(arr))
print(np.log10(arr))
print(np.exp(arr))
```

```python
# Rounding
print(np.round(10.7))
print(np.ceil(10.5))
print(np.floor(10.5))
print(np.trunc(10.5))
```

```python
# some other functions
arr = [1,2,3,4,1,7,-9,10,4,-2]
print(np.unique(arr))  #gives Unique Value
print(np.sort(arr))
print(np.abs(arr))
```
